```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading an Image

```
img = cv2.imread('/content/drive/My Drive/OpenCV/apple.jpg',cv2.IMREAD_GRAYSCALE )
```

Showing Image using Matplotlib

```
plt.imshow(img)
```

```
plt.imshow(img, cmap = 'hot')
```

```
plt.imshow(img, cmap = 'gray')
```

With interpolation

```
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
```

Plotting Images

```
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.plot([50, 100], [80, 100],'r', linewidth = 10)
```

Saving Images

```
cv2.imwrite('apple_copy.png', img)
```

```
img_copy = cv2.imread('/content/apple_copy.png', cv2.IMREAD_GRAYSCALE)
```
