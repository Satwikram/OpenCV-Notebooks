```
add3 = cv2.add(img/2, img2/2)
cv2_imshow(add3)
```

Adding weights to images

```
weighted = cv2.addWeighted(img, 0.6, img2, 0.4, 0) # here 0.6 is 60% of weight and 0.4 is 40%
cv2_imshow(weighted)
```
